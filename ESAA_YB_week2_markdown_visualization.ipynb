### 02 맷플롯립(Matplotlib)

01) 맷플롯립의 pyplot 모듈의 이해

* pyplot은 plot() 함수를 통해 선 그래프를 그릴 수 있음.
*  title() 함수는 입력값으로 타이틀 문자열을 받아서 차트 타이틀 설정 가능

````python
# pyplot 모듈을 (관용적으로) plt라는 이름으로 import.
import matplotlib.pyplot as plt

# x축 입력값 [1,2,3], y축 입력값 [2,4,6]으로 선그래프 생성.
plt.plot([1,2,3], [2,4,6])

# 선그래프 타이틀을 Hello plot으로 설정
plt.title("Hello plot")

# 선 그래프를 출력
plt.show()
````

3) Figure와 Axis의 활용

*  Figure은 주로 전체 그림판의 크기를 조절하는데 사용

````python
# plt.figure()은 주로 figure의 크기를 조절하는 데 사용.
# 아래는 figure 크기가 가로 10, 세로 4인 Figure 객체를 설정하고 반환
plt.figure(figsize=(10,4))
plt.plot([1,2,3],[2,4,6])
plt.title('Hello plot')
plt.show()
````

````python
figure = plt.figure(figsize=(10,4))
print(type(figure))
````

````python
# figure() 함수의 인자로 facecolor='yellow'를 통해 그림판의 배경색을 노란색으로 변경
plt.figure(figsize=(10,4), facecolor='yellow')
plt.plot([1,2,3],[2,4,6])
plt.title('Hello plot')
plt.show()
````

````python
ax = plt.axes()
print(type(ax))
````

````python
# pyplot에서 figure와 axes 객체를 함꼐 가져오기 - plt.subplots()이용
### pyplot에서 설정된 Figure와 Axes 객체를 함께 가져오기
fig, ax = plt.subplots()
print('fig type:', type(fig), '\nax type:', type(ax))
````

````python
fig, ax = plt.subplots(figsize=(4,4))
# Axes 객체를 바로 이용해 선그래프 그림.
ax.plot([1,2,3],[2,4,6])
# Axes 객체의 set_title()을 이용하여 선 그래프 타이틀을 Hello plot으로 설정
ax.set_title('Hello plot')
plt.show()
````

6) 축 명칭 설정, 축의 눈금(틱)값 회전, 범례(legend) 설정

*  축의 명칭을 설정하기 원한다면 pyplot의 xlabel(), ylabel()과 같은 함수를 이용해 축의 명칭을 설정

````python
plt.plot(x_value, y_value, color='red', marker='o', linestyle='dashed', linewidth=2, markersize=12)
# x축과 y축의 이름을 텍스트로 할당하여 설정
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
````

````python
# x 좌표 값을 0~99까지 100개로 설정.
x_value = np.arange(0,100)
# y 좌표 값은 개별 x좌표값을 2배로 해서 100개로 설정.
y_value = 2*x_value

plt.plot(x_value, y_value, color='green')
plt.xlabel('x axis')
plt.ylabel('y axis')

# x축 눈금값을 90도로 회전
plt.xticks(rotation=90)
plt.show()
````

* x 축의 눈금 값을 좀 더 세밀하게 나타내고 싶으면 xticks() 함수의 인자로 ticks 값을 나태내고 싶은 값을 설정

````python
x_value = np.arange(0,100)
y_value = 2*x_value

plt.plot(x_value, y_value, color='green')
plt.xlabel('x axis')
plt.ylabel('y axis')

# x축의 눈금값을 np.arange(0, 100, 5)로 하여 0부터 99까지 5개의 스텝으로 20개를 표시하고
# 90도 회전
plt.xticks(ticks=np.arange(0,100,5), rotation=90)
plt.show()
````

*  범례 설정: 범례는 여러 개의 그래프/차트를 한 Axes 내에서 표시할 때 편리하게 사용

````python
# x좌표값은 0~99까지 100개, 첫번쨰 Y좌표값은 X좌표값의 두 배, 두번째 좌표값은
# x좌표값의 4배
x_value_01 = np.arange(0,100)
y_value_01 = 2*x_value_01
y_value_02 = 4*x_value_01

# x좌표값과 첫번째 y값으로 녹색의 선그래프를 그림. label을 'green line'으로 설정.
plt.plot(x_value_01, y_value_01, color='green', label='green line')
# x좌표값과 두번쨰 y좌표값으로 빨간색 선 그래프를 그림. label을 'rend line'으로 설정.
plt.plot(x_value_01, y_value_02, color='red', label='red line')

plt.xlabel('x axis')
plt.ylabel('y axis')

# 개별 plot들의 label로 설정된 문자열을 범례로 표시
plt.legend()
plt.title('Hello plot')
plt.show()
````

````python
# 선과 막대 그래프 함께 표시
x_value_01 = np.arange(1,10)
y_value_01 = 2*x_value_01

# 마커를 포함한 빨간색 대시 선 그래프를 그리고, label은 'red line'으로 설정
plt.plot(x_value_01, y_value_01, color='red', marker='o', linestyle='dashed', label='red lines')
# x값에 따른 y값을 나타내는 초록색 막대 그래프를 그리고 label은 'bar plot'으로 설정
plt.bar(x_value_01, y_value_01, color='green', label='bar plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
# 두 개의 그래프에서 label 설정된 문자열 값을 범례로 표시
plt.legend()
plt.title('Hello plot')
plt.show()
````

````python
figure = plt.figure()
# Axes 객체를 추출하여 ax 변수에 할당
ax = plt.axes()

# plt.plot()은 ax.plot으로 plt.bar()은 ax.bar()로 변경
ax.plot(x_value_01, y_value_01, color='red', marker='o', linestyle='dashed', label='red lines')
ax.bar(x_value_01, y_value_01, color='green', label='bar plot')

#plt.xlabel()은 ax.set_xlabel()로, plt.ylabel()은 ax.set_ylabel()로 변경
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')

#ax.set_legend()가 아니라 ax.legend()임.
ax.legend()
# plt.title()을 plt.set_title()로 변경.
ax.set_title('Hello plot')
plt.show()
````

7) 여러 개의 subplots들을 이용해 개별 그래프들을 subplot별로 시각화 하기


````python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(6,3))
````

````python
print('ax type:', type(ax))
print('ax[0] type:', type(ax[0]))
print('ax[1] type:', type(ax[1]))
````

````python
import numpy as np

x_value_01 = np.arange(1,10)
x_value_02 = np.arange(1,20)
y_value_01 = 2 * x_value_01
y_value_02 = 4 * x_value_02

# figsize는 (6,3)인 Figure와 2개의 Axes를 가지는 subplots을 반환
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(6,3))

#개별 Axes 객체들을 각각 이용하여 선 그래프와 막대그래프를 그림.
ax[0].plot(x_value_01, y_value_01, color='red', marker='o', linestyle='dashed', label='red lines')
ax[1].bar(x_value_02, y_value_02, color ='green', label='bar plot')
# 개별 Axes 객체 각각에 범례를 설정
ax[0].legend()
ax[1].legend()
# 개별 Axes 객체 각각에 타이틀을 설정
ax[0].set_title('Hello line')
ax[1].set_title('Hello bar')
plt.show()
````

````python
import numpy as np

x_value_01 = np.arange(1,10)
x_value_02 = np.arange(1,20)
y_value_01 = 2 * x_value_01
y_value_02 = 4 * x_value_02

# figsize는 (6,6)인 Figure와 2*2개의 Axes를 가지는 subplots을 반환.
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(6,6))

# 2*2 Axes 객체들을 각각 이용하여 선 그래프와 막대 그래프를 그림.
ax[0][0].plot(x_value_01, y_value_01, color='red', marker='o', linestyle='dashed', label='red line')
ax[0][1].bar(x_value_02, y_value_02, color='green', label='bar plot')
ax[1][0].plot(x_value_01, y_value_01, color='green', marker='o', linestyle = 'dashed', label='green line')
ax[1][1].bar(x_value_02, y_value_02, color='red', label='red bar')

# 개별 Axes 객체 각각에 X축명을 설정.
ax[0][0].set_xlabel('x axis')
ax[0][1].set_xlabel('x axis')
ax[1][0].set_xlabel('x axis')
ax[1][1].set_xlabel('x axis')

# 개별 Axes 객체 각각에 범례를 설정
ax[0][0].legend()
ax[0][1].legend()
ax[1][0].legend()
ax[1][1].legend()

plt.show()
````
