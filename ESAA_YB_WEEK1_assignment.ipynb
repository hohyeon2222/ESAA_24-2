#### 19. Normalize a 5x5 random matrix (★☆☆)
Hint : $Z = \frac{X - \overline{X}}{S}$

````python
x = np.random.random((5,5))
print('random matrix:\n', x)
x_bar = x.mean()
s = x.std()
z = (x - x_bar)/s
print('normalized matrix:\n',z)
````
